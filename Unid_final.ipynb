rfc 구현 및 학습

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score

model3 = RandomForestClassifier(random_state = 1, n_jobs = 10, n_estimators = 1000, min_samples_split=40, max_depth=100, bootstrap=False)
model3.fit(train_x, train_y)

val_pred_1 = model3.predict(val_x)
print("Validation F1 score: ", f1_score(val_y, val_pred_1, average="macro"))
```

kkn 구현 및 학습

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score

classifier = KNeighborsClassifier(n_neighbors = 5, weights="distance", metric="manhattan")
classifier.fit(train_x, train_y)

val_pred_knn = classifier.predict(val_x)
print("Validation F1 score: ", f1_score(val_y, val_pred_knn, average="macro"))
```

모델 합치는 부분 - voting model 뒤에 클래시파이어 추가

```python
from sklearn.ensemble import VotingClassifier

voting_model = VotingClassifier(estimators=[('Random Forest', model3), ('Knn', classifier)], voting='soft') # voting->hard, soft
```

이거 실행하면 학습 후 validation에 대해 f1 score 출력함 -> voting_model로 test 파일에 대해 예측해서 제출용 파일 만들면 끝

```python
voting_model.fit(train_x, train_y)
pred1 = voting_model.predict(val_x)
print("F1 score: ", f1_score(val_y, pred1, average="macro"))
```
